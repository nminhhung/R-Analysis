```r
install.packages("XML")
```

```r
library('XML')
```

```r
# create a new xml doc
doc_xml <- newXMLDoc(isHTML = FALSE)

# create a table node
table_node <- newXMLNode("table", doc = doc_xml)

# row data
row_data <- apply(df_golden_delicious, 1, function(x) {
  z1 <- newXMLNode('row') # create a new node for each row
  addChildren(z1, lapply(names(x), function(y) newXMLNode(y, x[y])))
})

# add row data to table node
xmlParent(row_data) <- table_node

# save as xml file
saveXML(doc_xml, file = "df_golden_delicious.xml")
```

```r
subtype_summarize = df %>% 
 group_by(fruit_subtype) %>% 
 summarise(
   freq = na_if(sum(!is.na(fruit_subtype)), 0),
   across(c(mass, width, height), list(avg = ~ ifelse(all(is.na(.x)), NA, mean(.x, na.rm = TRUE)), 
                                   min = ~ ifelse(all(is.na(.x)), NA, min(.x, na.rm = TRUE)), 
                                   max = ~ ifelse(all(is.na(.x)), NA, max(.x, na.rm = TRUE))))
   )
```
